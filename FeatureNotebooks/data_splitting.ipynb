```python
import pandas as pd
import numpy as np
np.random.seed(42)
```

```python
# splitting data by lesion_id - without data augmentation and oversampling
metadata_df = pd.read_csv('/Users/joanna/datasci_281/final_project/HAM10000_metadata.csv')
metadata_df.head()
```

```python
unique_lesion_id = metadata_df['lesion_id'].unique()
len(unique_lesion_id)
```

```python
# splitting data into 70% training, 15% validation, 15% testing
np.random.shuffle(unique_lesion_id)
train_ind = round(len(unique_lesion_id) * 0.70)
validation_ind = round(len(unique_lesion_id) * 0.85)

train_lesion_id = unique_lesion_id[:train_ind]
val_lesion_id = unique_lesion_id[train_ind:validation_ind]
test_lesion_id = unique_lesion_id[validation_ind:]
```

```python
train_df = metadata_df[metadata_df['lesion_id'].isin(train_lesion_id)]
val_df = metadata_df[metadata_df['lesion_id'].isin(val_lesion_id)]
test_df = train_df = metadata_df[metadata_df['lesion_id'].isin(test_lesion_id)]
```

```python
features_df = pd.read_csv('/Users/joanna/datasci_281/final_project/dataset/dataset.csv')
features_df = features_df.rename(columns = {'Unnamed: 0': 'image_id'})
```

```python
train_image_id = train_df['image_id']
train_features_df = features_df[features_df['image_id'].isin(train_image_id)]
train_Y_df = train_df[['image_id', 'dx']]
train_features_df = pd.merge(train_features_df, train_Y_df, on = 'image_id')
train_features_df.set_index('image_id', inplace = True)

val_image_id = val_df['image_id']
val_features_df = features_df[features_df['image_id'].isin(val_image_id)]
val_Y_df = val_df[['image_id', 'dx']]
val_features_df = pd.merge(val_features_df, val_Y_df, on = 'image_id')
val_features_df.set_index('image_id', inplace = True)

test_image_id = test_df['image_id']
test_features_df = features_df[features_df['image_id'].isin(test_image_id)]
test_Y_df = test_df[['image_id', 'dx']]
test_features_df = pd.merge(test_features_df, test_Y_df, on = 'image_id')
test_features_df.set_index('image_id', inplace = True)
```
